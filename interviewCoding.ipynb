Factorial of a Number

```python
def fac(n):
    temp=1
    for i in range(1,n+1):
        temp=temp*i
    return temp

n=5
print(fac(n))
```

Check Prime Number

```python
def prime(n):
    if n<2:
        return False
    for i in range(2,n):
        if n%i==0:
            return False
    return True


pr=[]
for i in range(2,100):
    if prime(i) is True:
        pr.append(i)
print(pr)
```

Fibonacci Series

```python
def fib(n):
    a,b=0,1
    for i in range(n):
        print(a, end=" ")
        a,b=b,a+b

n=int(input("enter a number:"))
fib(n)
```

Reverse a Number (without using string)

```python
def reverse(n):
    ans=0
    while n!=0:
        rem=n%10
        ans=ans*10+rem
        n=n//10
    return ans

print(reverse(1234))
```

Check if a number is Palindrome

```python
def reverse(n):
    ans=0
    while n!=0:
        rem=n%10
        ans=ans*10+rem
        n=n//10
    return ans

n=int(input())
temp=n
if temp==reverse(n):
    print("Palindrome")
else:
    print("no Palindrome")
```

Find GCD of Two Numbers

```python
def GCD(f,s):
    ans=1
    for i in range(1,min(f,s)):
        if f%i==0 and s%i==0:
            ans=i
    return ans

f=12
s=18
print(GCD(f,s))
```

```python
def GCD(n1,n2):
    while n2!=0:
        n1,n2=n2,n1%n2
    return n1

print(GCD(36,60))
```

Find Largest Element in an Array

```python
def lar(li):
    ans=li[0]
    for i in li:
        if i>ans:
            ans=i
    return ans

li=[10, 45, 23, 89, 12]
print(lar(li))
```

Find Second Largest Element

```python
def sec(li):
    fir=0
    sec=0
    for i in li:
        if i>fir:
            sec=fir
            fir=i
        elif fir>i>sec:
            sec=i
    return sec


li=[10, 45, 23, 89, 12]
print(sec(li))
```
