4️⃣ Largest Sum Subarray (Maximum Subarray Sum)

Problem:
Given an array of integers (can contain negative numbers), find the maximum possible sum of any contiguous subarray.

(This is the classic Kadane’s Algorithm problem.)

Input

First line: integer n

Second line: n space-separated integers

Output

Single integer → maximum subarray sum.

```python
def max_subarray_sum(arr):
    # assume arr has at least one element
    curr_sum = arr[0]
    max_sum = arr[0]

    for x in arr[1:]:
        # best sum ending at current position
        curr_sum = max(x, curr_sum + x)
        # update overall best
        max_sum = max(max_sum, curr_sum)

    return max_sum


# Input handling
n = int(input())
arr = list(map(int, input().split()))

print(max_subarray_sum(arr))
```
